# 06 - Proyecto de ejemplo: análisis rápido de evento

En este notebook integramos lectura, cálculo y mapa.

```
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

ds = xr.open_dataset('../data/wrfout_sample.nc')
T2 = ds['T2'].isel(time=0) - 273.15  # C
wind = np.sqrt(ds['U10']**2 + ds['V10']**2).isel(time=0)
lon = ds['lon'].values
lat = ds['lat'].values
lon2d, lat2d = np.meshgrid(lon, lat)

fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection=ccrs.PlateCarree())
cf = ax.contourf(lon2d, lat2d, T2, transform=ccrs.PlateCarree(), cmap='coolwarm')
q = ax.quiver(lon2d, lat2d, ds['U10'].isel(time=0), ds['V10'].isel(time=0), transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_title('Proyecto: Temperatura (°C) y viento (m/s) - sintético')
plt.colorbar(cf, orientation='vertical', pad=0.02)
plt.show()
```

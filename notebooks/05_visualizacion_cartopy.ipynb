# 05 - Visualización con Cartopy

Mapas simples y superposición de campos.

```
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import xarray as xr
import numpy as np

ds = xr.open_dataset('../data/wrfout_sample.nc')
T2 = ds['T2'].isel(time=0)
lon = ds['lon'].values
lat = ds['lat'].values
lon2d, lat2d = np.meshgrid(lon, lat)

fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection=ccrs.PlateCarree())
cs = ax.contourf(lon2d, lat2d, T2, transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_title('Temperatura 2m (K) - ejemplo sintético')
plt.colorbar(cs, orientation='horizontal', pad=0.05)
plt.show()
```

**Nota:** en Colab puede ser necesario instalar `cartopy` antes de ejecutar esta celda.

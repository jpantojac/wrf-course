```
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import xarray as xr
import numpy as np

url = "https://github.com/jpantojac/wrf-course/blob/main/data/wrfout_d01_2025-11-12_17-00-00"

!wget -O wrfout_d01_2025-11-12_17-00-00.nc $url

import xarray as xr
ds = xr.open_dataset('wrfout_d01_2025-11-12_17-00-00.nc')

T2 = ds['T2'].isel(time=0)
lon = ds['lon'].values
lat = ds['lat'].values
lon2d, lat2d = np.meshgrid(lon, lat)

fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection=ccrs.PlateCarree())
cs = ax.contourf(lon2d, lat2d, T2, transform=ccrs.PlateCarree())
ax.coastlines()
ax.set_title('Temperatura 2m (K) - ejemplo sintético')
plt.colorbar(cs, orientation='horizontal', pad=0.05)
plt.show()
```
